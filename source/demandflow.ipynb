```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

```

```python

# Create figure and axis
fig, ax = plt.subplots(figsize=(12,8))

# Define boxes (process steps)
boxes = [
    ("Aggregate Forecast\n(Product Family A = 10,000 units)", (0.05, 0.8)),
    ("Disaggregation / Allocation\n- Plant 1: 6,000\n- Plant 2: 4,000", (0.35, 0.8)),
    ("Rough-Cut Capacity Planning (RCCP)\n- Plant 1 cap: 5,500  Bottleneck\n- Plant 2 cap: 5,500  OK", (0.65, 0.8)),
    ("Detailed Supply Plan\n- Plant 1: 5,500\n- Plant 2: 4,500", (0.35, 0.55)),
    ("Execution & Forecast Consumption\n- Orders: P1=5,200 | P2=4,700", (0.35, 0.3)),
    ("Inventory Positioning\n- Plant 1  DC East (3,000)\n- Plant 1  DC Central (2,200)\n- Plant 2  DC West (2,500)\n- Plant 2  DC Central (2,200)", (0.35, 0.05))
]

# Draw boxes
for text, (x, y) in boxes:
    ax.add_patch(mpatches.FancyBboxPatch((x, y), 0.28, 0.13, boxstyle="round,pad=0.02",
                                         fc="lightgreen" if "Inventory" in text else "lightyellow",
                                         ec="black", lw=1.5))
    ax.text(x+0.14, y+0.065, text, ha="center", va="center", fontsize=9, wrap=True)

# Draw arrows
arrow_props = dict(arrowstyle="->", lw=1.5, color="black")
ax.annotate("", xy=(0.33, 0.865), xytext=(0.05+0.28, 0.865), arrowprops=arrow_props)
ax.annotate("", xy=(0.63, 0.865), xytext=(0.35+0.28, 0.865), arrowprops=arrow_props)
ax.annotate("", xy=(0.49, 0.62), xytext=(0.49, 0.8), arrowprops=arrow_props)
ax.annotate("", xy=(0.49, 0.37), xytext=(0.49, 0.55), arrowprops=arrow_props)
ax.annotate("", xy=(0.49, 0.12), xytext=(0.49, 0.3), arrowprops=arrow_props)

# Hide axes
ax.axis("off")

plt.title("Demand Flow with Inventory Positioning to Distribution Centers", fontsize=12, weight="bold")
plt.show()
```
